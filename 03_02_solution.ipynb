# Parsing datetime data

What is the most landslide-prone month in a province or state? Extract data and show the distribution.

<strong>Challenge:</strong>
<ul>
<li>Convert the date column into a DateTime data type and extract month from the date to visualize their frequencies using a histogram or KDE plot.</li>
</ul>

<strong>Tips:</strong>
<ul>
<li>Use pd.to_datetime() function to convert to datetime object</li>
<li>Extract month using .dt.month attribute</li>
<li>Use seaborn or the matplotlib library to visualize the distribution using KDE plot or histogram</li>
</ul>

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/landslides.csv")
df.head()
```

## Check and handle missing values

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df[~df['date'].isnull()]
```

```python
df
```

```python
df['time'].value_counts()
```

```python
df['time'].isna().sum()
```

```python
df['time'] = df['time'].fillna("Not Known")
df
```

```python
df['time'].isna().sum()
```

```python
mean = df['fatalities'].mean()
mean
```

```python
df['fatalities'] = df['fatalities'].fillna(mean)
df
```

```python
df['fatalities'].isna().sum()
```

## Date time parsing using .to_datetime()

```python
df["date"] = pd.to_datetime(df["date"], format="%m/%d/%y")
df
```

```python
df["month"] = df["date"].dt.month
month = df.pop("month")
df.insert(2, "month", month)
df
```

```python
df["month"].value_counts()
```

```python
import seaborn as sns

sns.set_style("darkgrid")
sns.histplot(data=df, x="month", kde=False, bins=12);
```
